```python
filmes = []
anos = []
index = []

n = True
while n ==True:
  filme = input("\n Informe o nome do Filme: ")
  ano = int(input("Informe o ano do Filme: "))
  filmes.append(filme)
  anos.append(ano)

  escolha = input("Deseja Continuar [S/N]: ").lower()
  if escolha == 'nao' or escolha == 'n':
    n = False
  
for i in range(0,len(filmes),1):
  index.append('Line '+ str(i+1))

data = {'Filmes':filmes, 'Ano': ano}
df2 = pd.DataFrame(data,index = index)
display(df2)
```
