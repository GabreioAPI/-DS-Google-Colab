```python
import pandas as pd
link = r'https://raw.githubusercontent.com/rafaelpuyau/infinity_school/main/ds/datasets/wines.csv'

df = pd.read_csv(link)
```

```python
df.sample(5)
```

```python
#1,2, 6 questao
preco_caro = df['price'].max()
preco_barato = df['price'].min()
quantidade = df['quantity'].sum()
```

```python
#3 e 4 questao
vinho_caro = df.loc[df['price'] == preco_caro,"label"].values
vinho_barato = df.loc[df['price'] == preco_barato,"label"].values
```

```python
#5 questao
df2 = df.loc[(df['date']>='2021-01-01') & (df['date']<='2021-12-31')]
total_vendidos = df2['quantity'].sum()
```

```python
#7 e 8 questao
rate_max = df['rate'].max()
rate_min = df['rate'].min()

vinho_maior_nota = df.loc[df['rate']==rate_max,"label"].drop_duplicates().values
vinho_menor_nota = df.loc[df['rate']==rate_min,"label"].drop_duplicates().values
```

```python
#9 e 10 questao
safra_antiga_data = df['vintage'].min()
safra_recente_data = df['vintage'].max()
```

```python
#11 e 12 questao
paises = df.groupby('country')['quantity'].sum().sort_values()
pais_min = paises.index[0]
pais_max = paises.index[-1]
total_max = paises.head(1).values
total_min = paises.tail(1).values
```

```python
#13 e 14 questao
df2 = df.loc[(df['price']<100)]
df3 = df.loc[(df['price']>100)]

abaixo_100 = df2['quantity'].sum()
acima_100 = df3['quantity'].sum()
```

```python
#15 e 16 questao
mes_dezembro = df.loc[(df['style'] == 'Sparkling') & (df['date']>='2021-12-01') & (df['date']<='2022-01-01')]
quantidade_dezembro = mes_dezembro['quantity'].sum()

outros_meses = df.loc[((df['style'] == 'White') | (df['style'] == 'Rose')) & (df['date']>='2021-01-01') & (df['date']<='2021-03-31')]
quantidade_rose_branco = outros_meses['quantity'].sum()
```

```python
print(f'1 - R${preco_caro}')
print(f'\n2 - R${preco_barato}')
print(f'\n3 - Vinho = {vinho_caro}')
print(f'\n4 - Vinho = {vinho_barato}')
print(f'\n5 - {total_vendidos} vinhos foram vendidos em 2021')
print(f'\n6 - {quantidade} garrafas no total')
print(f'\n7 - Vinhos avaliados com nota maxima = {vinho_maior_nota}')
print(f'\n8 - Vinhos avaliados com nota minima = {vinho_menor_nota}')
print(f'\n9 - Safra mais antiga = {safra_antiga_data}')
print(f'\n10 - Safra mais recente = {safra_recente_data}')
print(f'\n11 - {pais_max} vendeu mais, com total de {total_max[0]} garrafas vendidas')
print(f'\n12 - {pais_min} vendeu menos, com total de {total_min[0]} garrafas vendidas')
print(f'\n13 - {abaixo_100} vinhos abaixo de R$100 foram vendidos')
print(f'\n14 - {acima_100} vinhos acima de R$100 foram vendidos')
print(f'\n15 - {quantidade_dezembro} vinhos foram vendidos em dezembro')
print(f'\n16 - {quantidade_rose_branco} vinhos roses e branco foram vendidos entre os meses de janeiro a março')
```
